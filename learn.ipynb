## 1. CLI 与 GUI

**CLI**（Command Line Interface）— 命令行界面  
**GUI**（Graphical User Interface）— 图形用户界面

| 对比项 | CLI | GUI |
|--------|-----|-----|
| 交互方式 | 键入文本命令 | 鼠标点击图形元素 |
| 学习曲线 | 较陡，需记忆命令 | 较平缓，所见即所得 |
| 操作效率 | 熟练后极高，支持脚本自动化 | 直觉友好，但重复操作效率较低 |
| 典型例子 | Terminal、PowerShell、Bash | Windows 资源管理器、VS Code |

## 2. Node.js 与 npm

- **Node.js** — JavaScript 的运行环境（让 JS 脱离浏览器在服务端运行）
- **npm**（Node Package Manager）— Node.js 的包管理器

### 与 Python 生态的类比

| JavaScript 生态 | Python 生态 | 作用 |
|-----------------|-------------|------|
| Node.js | Python 解释器 | 运行代码的环境 |
| npm | pip | 安装和管理第三方包 |
| `package.json` | `requirements.txt` | 记录项目依赖 |

## 3. Claude CLI 安装指南（Windows）

### 前置依赖
- Node.js & npm
- Git

### 安装步骤

**第一步：安装 Claude Code**
```bash
npm install -g @anthropic-ai/claude-code
```

**第二步：验证安装**
```bash
claude --version
# 输出示例：2.1.31 (Claude Code)
```

> **提示**：安装完成后，在任意项目目录下运行 `claude` 即可启动交互式 AI 编程助手。

## 4. Agent 记忆文件

Claude Code 在每个工作区（workspace）下维护一个 `.claude` 文件夹，用于存储轻量级的上下文记忆。

### 特点
- **位置**：项目根目录下的 `.claude/` 文件夹
- **内容**：记录用户偏好、权限设置等基础配置
- **范围**：记忆量较小，仅保存最基本的工作区上下文，不会存储大量对话历史

> **注意**：`.claude/settings.local.json` 中包含本地权限配置（如允许执行的 bash 命令），通常不应提交到版本控制中。

## 5. MCP 与 Skill 管理

**MCP**（Model Context Protocol）— 让 Claude 连接外部工具和数据源的协议。

### 查看已安装的 MCP

```bash
# 方式一：命令行
claude mcp list

# 方式二：在 Claude CLI 交互界面中
/mcp
```

### 安装 MCP

```bash
# 示例：安装 GitHub MCP Server
claude mcp add github --scope user -- npx -y @modelcontextprotocol/server-github
```

### 删除 MCP

```bash
claude mcp remove github
```
